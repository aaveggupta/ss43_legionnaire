```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
%matplotlib inline
```

```python
import pickle
```

```python
tempdata=pd.read_csv('training 3.csv')
```

```python
tempdata
```

```python
tempdata.drop('Weight',axis=1,inplace=True)
```

```python
tempdata
```

```python
newdataset=tempdata['Target']
```

```python
newdataset=pd.get_dummies(newdataset)
```

```python
newdataset['fever']
```

```python
newdataset['prognosis']=tempdata['Source']
```

```python
newdataset
```

```python
dataframe1=pd.read_csv('Training.csv')
dataframe2=pd.read_csv('Training2.csv')
```

```python
finaldataset=pd.concat([dataframe1, dataframe2,newdataset], ignore_index=True,sort=False)
```

```python
set(finaldataset['prognosis'])
```

```python
finaldataset=finaldataset[finaldataset.prognosis!='suicide attempt']
```

```python
finaldataset.fillna(0,inplace=True)
```

```python
finaldataset
```

```python
len(finaldataset)
```

```python
labels=finaldataset['prognosis']
fdc=finaldataset
fdc.drop('prognosis',axis=1,inplace=True)
```

```python
features_list=list(fdc.columns)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(fdc,labels,test_size=0.25,random_state=20)
```

```python
len(features_list)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
symptom=np.zeros([526],dtype=float)
l=mnb.classes_.tolist()
user_symptoms = ['shivering','blackheads','weightgain']
m=[]
for i in range(len(x_train.columns)):
    if x_train.columns[i] in user_symptoms:
        m.append(i)
for i in m:
    symptom[i]=1
probab=mnb.predict_proba([symptom]).tolist()
top5=[]
for j in range(5):
    max=-100000000
    h=0
    for i in range(len(probab[0])):
        if probab[0][i]>max:
            max=probab[0][i]
            h=i 
    k=[]
    k.append(l[h])
    k.append(probab[0][h])
    top5.append(k)
    probab[0][h]=-1    
```
